# 固有値と固有ベクトルの計算

````python
import pandas as pd
import numpy as np

# サンプルのデータフレーム（対称行列）
df = pd.DataFrame([[2, 1], 
                   [1, 2]])

# NumPy配列に変換
matrix = df.values

# 固有値と固有ベクトルの計算
eigvals, eigvecs = np.linalg.eig(matrix)

print("固有値:", eigvals)
print("固有ベクトル:\n", eigvecs)
````

# 回転行列


````python
import numpy as np
import matplotlib.pyplot as plt

N = 100
theta = np.linspace(-np.pi, np.pi, N+1)
p_before = np.column_stack((np.cos(theta), np.sin(theta)))

plt.plot(p_before[:,0], p_before[:,1], 'o')
plt.axis('equal')
plt.title("Before")
plt.show()

A = np.array([[1, 2],
              [3, 4]])
p_after = p_before @ A.T

plt.plot(p_after[:,0], p_after[:,1], 'o')
plt.axis('equal')
plt.title("After")
plt.show()
````

## コードの説明

### 1. 円周上の点を生成

```python
theta = np.linspace(-np.pi, np.pi, N+1)
p_before = np.column_stack((np.cos(theta), np.sin(theta)))
```

* `theta` は $-\pi$ から $\pi$ までを等分した角度の配列。
* $(\cos \theta, \sin \theta)$ を計算して、単位円周上の点を作成。

---

### 2. 単位円を描画

```python
plt.plot(p_before[:,0], p_before[:,1], 'o')
```

* `p_before` の点をプロットすると、単位円が表示される。

---

### 3. 行列による線形変換

```python
A = np.array([[1, 2],
              [3, 4]])
p_after = p_before @ A.T
```

* 2×2 の行列 `A` を用意。
* 単位円の座標 `p_before` に対して、`A` を掛けて座標変換を行う。
* `@` は行列の積

---

### 4. 変換後の点を描画

```python
plt.plot(p_after[:,0], p_after[:,1], 'o')
```

* 行列変換後の点を描画。
* 単位円は「楕円」や「傾いた形」に変換される。

---

## まとめ

* **最初**：単位円の点を生成して描画。
* **その後**：行列 `A` による線形変換を適用し、円が別の図形に変形する様子を描画。
