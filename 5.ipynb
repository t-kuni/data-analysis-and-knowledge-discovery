# üü¶ Ê≠£Ë¶èÂàÜÂ∏É (Normal distribution)

* Âπ≥Âùá„Çí‰∏≠ÂøÉ„Å´Â∑¶Âè≥ÂØæÁß∞„ÅÆÈá£ÈêòÂûã„ÅÆÂàÜÂ∏É„ÄÇËá™ÁÑ∂ÁèæË±°„ÇÑÊ∏¨ÂÆöË™§Â∑Æ„Å™„Å©„ÄÅÂ§ö„Åè„ÅÆ‰∫ãË±°„ÅßÁèæ„Çå„ÇãÂü∫Êú¨ÁöÑ„Å™ÂàÜÂ∏É„ÄÇ
* ÈÄ£Á∂ö

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# „Éë„É©„É°„Éº„Çø
mu = 0      # Âπ≥ÂùáŒº
sigma = 1   # Ê®ôÊ∫ñÂÅèÂ∑ÆœÉ

# xËª∏„ÅÆÁØÑÂõ≤
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Ê≠£Ë¶èÂàÜÂ∏É„ÅÆÁ¢∫ÁéáÂØÜÂ∫¶Èñ¢Êï∞
y = norm.pdf(x, mu, sigma)

# „Ç∞„É©„ÉïÊèèÁîª
plt.plot(x, y, label=f'N({mu}, {sigma**2})')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

### np.linspace

`np.linspace(start, stop, num)` „ÅØ„ÄÅÊåáÂÆö„Åó„ÅüÂå∫Èñì `[start, stop]` „ÇíÁ≠âÈñìÈöî„Å´ÂàÜÂâ≤„Åó„ÅüÊï∞Âàó„ÇíÁîüÊàê„Åô„ÇãÈñ¢Êï∞„ÄÇ
„Åì„Åì„Åß„ÅØ `mu - 4*sigma` „Åã„Çâ `mu + 4*sigma` „Åæ„Åß„Çí1000ÂàÜÂâ≤„Åó„ÅüÈÖçÂàó„Çí‰Ωú„Çä„ÄÅ„Ç∞„É©„Éï„ÅÆxËª∏„ÅÆÂÄ§„Å®„Åó„Å¶‰Ωø„Å£„Å¶„ÅÑ„Çã„ÄÇ

---

### norm.pdf

`scipy.stats.norm.pdf(x, loc, scale)` „ÅØÊ≠£Ë¶èÂàÜÂ∏É„ÅÆÁ¢∫ÁéáÂØÜÂ∫¶Èñ¢Êï∞ (Probability Density Function) „ÅÆÂÄ§„ÇíËøî„ÅôÈñ¢Êï∞„ÄÇ

* `x`: Ë©ï‰æ°„Åô„ÇãÁÇπ
* `loc`: Âπ≥ÂùáÂÄ§Ôºà„Åì„Åì„Åß„ÅØ `mu`Ôºâ
* `scale`: Ê®ôÊ∫ñÂÅèÂ∑ÆÔºà„Åì„Åì„Åß„ÅØ `sigma`Ôºâ

ÁµêÊûú„Å®„Åó„Å¶„ÄÅÂêÑ `x` „Å´ÂØæÂøú„Åô„ÇãÊ≠£Ë¶èÂàÜÂ∏É„ÅÆÈ´ò„Åï„ÇíË®àÁÆó„Åó„Å¶„ÅÑ„Çã„ÄÇ

---

### label=f'N({mu}, {sigma**2})'

Python „ÅÆ **fÊñáÂ≠óÂàó (f-string)** „Åß„ÄÅÂ§âÊï∞„ÇíÊñáÂ≠óÂàó„Å´Âüã„ÇÅËæº„Çì„Åß„ÅÑ„Çã„ÄÇ

* `mu` ‚Üí Âπ≥Âùá
* `sigma**2` ‚Üí ÂàÜÊï£ÔºàÊ®ôÊ∫ñÂÅèÂ∑Æ„ÅÆ2‰πóÔºâ

„Åì„ÅÆ„Ç≥„Éº„Éâ„Åß„ÅØÂá°‰æã„Å´ `N(Âπ≥Âùá, ÂàÜÊï£)` „ÅÆÂΩ¢„ÇíËá™ÂãïÁöÑ„Å´Ë°®Á§∫„Åô„Çã„ÄÇ
‰æã„Åà„Å∞ `mu=0, sigma=1` „Å™„Çâ `"N(0, 1)"` „Åå„É©„Éô„É´„Å´„Å™„Çã„ÄÇ



# üü¶ ‰∫åÈ†ÖÂàÜÂ∏É (Binomial distribution)

* ‰∏ÄÂÆö„ÅÆË©¶Ë°åÂõûÊï∞„ÅÆ‰∏≠„Åß„ÄÅÊàêÂäü„ÉªÂ§±Êïó„ÅÆ‰∫åÂÄ§„ÅåËµ∑„Åì„ÇãÁ¢∫Áéá„ÇíË°®„ÅôÂàÜÂ∏É„ÄÇ‰æã„Åà„Å∞„Äå„Ç≥„Ç§„É≥„Çí10ÂõûÊäï„Åí„Å¶Ë°®„ÅåÂá∫„ÇãÂõûÊï∞„Äç„ÄÇ
* Èõ¢Êï£

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# „Éë„É©„É°„Éº„Çø
n = 20   # Ë©¶Ë°åÂõûÊï∞
p = 0.5  # ÊàêÂäüÁ¢∫Áéá

# x„ÅÆÁØÑÂõ≤
x = np.arange(0, n+1)

# Á¢∫ÁéáË≥™ÈáèÈñ¢Êï∞ (PMF)
pmf = binom.pmf(x, n, p)

# „Ç∞„É©„ÉïÊèèÁîª
plt.bar(x, pmf, tick_label=x)
plt.xlabel("ÊàêÂäüÂõûÊï∞")
plt.ylabel("Á¢∫Áéá")
plt.title(f"‰∫åÈ†ÖÂàÜÂ∏É (n={n}, p={p})")
plt.show()
```

* `pmf` „ÅØ Á¢∫ÁéáË≥™ÈáèÈñ¢Êï∞ (probability mass function) „ÅÆÁï•„Åß„Åô„ÄÇ

# üü¶ Âπæ‰ΩïÂàÜÂ∏É(Geometric distribution)

* Èõ¢Êï£
* ÊúÄÂàù„Å´ÊàêÂäü„ÅåÂá∫„Çã„Åæ„Åß„ÅÆË©¶Ë°åÂõûÊï∞„ÇíË°®„ÅôÂàÜÂ∏É„ÄÇ‰æã„Åà„Å∞„Äå„Ç≥„Ç§„É≥„ÇíÊäï„Åí„Å¶Âàù„ÇÅ„Å¶Ë°®„ÅåÂá∫„Çã„Åæ„Åß„Å´„Åã„Åã„ÇãÂõûÊï∞„Äç„ÄÇ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import geom

# „Éë„É©„É°„Éº„Çø
p = 0.3 # ÊàêÂäüÁ¢∫Áéá

k = np.arange(1, 15)
pmf = geom.pmf(k, p)

plt.bar(k, pmf)
plt.xlabel("Ë©¶Ë°åÂõûÊï∞ (k)")
plt.ylabel("Á¢∫ÁéáË≥™ÈáèÈñ¢Êï∞ P(X=k)")
plt.title(f"Âπæ‰ΩïÂàÜÂ∏É (p={p})")
plt.show()
```

# üü¶ „Éù„Ç¢„ÇΩ„É≥ÂàÜÂ∏É(Poisson distribution)

* ‰∏ÄÂÆöÊôÇÈñìÂÜÖ„Åß„Åæ„Çå„Å´Ëµ∑„Åì„Çã‰∫ãË±°„ÅÆÂõûÊï∞„ÇíË°®„ÅôÂàÜÂ∏É„ÄÇ‰æã„Åà„Å∞„Äå1ÊôÇÈñì„Å´Ëµ∑„Åì„Çã‰∫§ÈÄö‰∫ãÊïÖ„ÅÆ‰ª∂Êï∞„Äç„ÄÇ
* Âçò‰ΩçÊôÇÈñìÂÜÖ„Å´Âπ≥ÂùáŒªÂõûÁô∫Áîü„Åô„Çã‰∫ãË±°„ÅåkÂõûÁô∫Áîü„Åô„ÇãÁ¢∫Áéá„ÅÆÂàÜÂ∏É„ÄÇÂΩìÁÑ∂ŒªÂõûÁô∫Áîü„Åô„Çã„Ç±„Éº„Çπ„Åå‰∏ÄÁï™Á¢∫Áéá„ÅåÈ´ò„ÅÑ„ÄÇ
* Èõ¢Êï£

$ \Huge P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0,1,2,\dots $

$ \Huge E[X] = \lambda $

$ \Huge \mathrm{Var}(X) = \lambda $


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# „Éë„É©„É°„Éº„Çø
lam = 4 # Âπ≥ÂùáÁô∫ÁîüÂõûÊï∞ (Œª)

# x Ëª∏„ÅÆÁØÑÂõ≤
x = np.arange(0, 15)

# „Éù„Ç¢„ÇΩ„É≥ÂàÜÂ∏É„ÅÆÁ¢∫ÁéáË≥™ÈáèÈñ¢Êï∞
pmf = poisson.pmf(x, lam)

# Ê£í„Ç∞„É©„Éï„ÅßÊèèÁîª
plt.bar(x, pmf)
plt.title(f"Poisson Distribution (Œª={lam})")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.show()
```

# üü¶ ‰∏ÄÊßòÂàÜÂ∏É (Uniform distribution)

* „ÅÇ„ÇãÁØÑÂõ≤ÂÜÖ„ÅÆÂÄ§„ÅåÁ≠â„Åó„ÅÑÁ¢∫Áéá„ÅßÂá∫Áèæ„Åô„ÇãÂàÜÂ∏É„ÄÇ‰æã„Åà„Å∞„Äå‰π±Êï∞„Åß0„Åã„Çâ1„Çí‰∏ÄÊßò„Å´ÈÅ∏„Å∂„Äç„ÄÇ
* ÈÄ£Á∂ö

```python
import numpy as np
import matplotlib.pyplot as plt

# „Éë„É©„É°„Éº„Çø
a, b = 0, 1 # ‰∏ÄÊßòÂàÜÂ∏É„ÅÆÁØÑÂõ≤

# „Éá„Éº„ÇøÁîüÊàê
x = np.linspace(a - 0.5, b + 0.5, 1000)
y = np.where((x >= a) & (x <= b), 1 / (b - a), 0)

# „Éó„É≠„ÉÉ„Éà
plt.plot(x, y, label=f'Uniform({a}, {b})')
plt.xlabel('x')
plt.ylabel('Á¢∫ÁéáÂØÜÂ∫¶')
plt.title('‰∏ÄÊßòÂàÜÂ∏É')
plt.legend()
plt.grid(True)
plt.show()
```

# üü¶ ÊåáÊï∞ÂàÜÂ∏É (Exponential distribution)

* „Éù„Ç¢„ÇΩ„É≥ÈÅéÁ®ã„Å´„Åä„Åë„Çã‰∫ãË±°„ÅÆÁô∫ÁîüÈñìÈöî„ÇíË°®„ÅôÂàÜÂ∏É„ÄÇ‰æã„Åà„Å∞„Äå„Éê„Çπ„ÅåÊù•„Çã„Åæ„Åß„ÅÆÂæÖ„Å°ÊôÇÈñì„Äç„ÄÇ
* ÈÄ£Á∂ö

```python
import numpy as np
import matplotlib.pyplot as plt

# „Éë„É©„É°„Éº„Çø
lam = 1.5 # Âπ≥ÂùáÁô∫ÁîüÂõûÊï∞ (Œª)

# xËª∏„ÅÆÁØÑÂõ≤
x = np.linspace(0, 5, 1000)

# ÊåáÊï∞ÂàÜÂ∏É„ÅÆÁ¢∫ÁéáÂØÜÂ∫¶Èñ¢Êï∞
pdf = lam * np.exp(-lam * x)

# „Ç∞„É©„ÉïÊèèÁîª
plt.plot(x, pdf, label=f'Œª={lam}')
plt.title("ÊåáÊï∞ÂàÜÂ∏É (Exponential Distribution)")
plt.xlabel("x")
plt.ylabel("Á¢∫ÁéáÂØÜÂ∫¶")
plt.legend()
plt.grid(True)
plt.show()
```

# üü¶ „Ç¨„É≥„ÉûÂàÜÂ∏É (Gamma distribution)

* Ë§áÊï∞„ÅÆÊåáÊï∞ÂàÜÂ∏É„ÅÆÂíå„Å®„Åó„Å¶Áèæ„Çå„ÇãÂàÜÂ∏É„ÄÇÂæÖ„Å°ÊôÇÈñì„ÅåÁ©ç„ÅøÈáç„Å™„ÇãÁä∂Ê≥Å„ÇíË°®Áèæ„Åß„Åç„Çã„ÄÇ
* ÈÄ£Á∂ö

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# „Éë„É©„É°„Éº„Çø
k = 2.0   # ÂΩ¢Áä∂ÊØçÊï∞
theta = 2.0  # Â∞∫Â∫¶ÊØçÊï∞

x = np.linspace(0, 20, 1000)
y = gamma.pdf(x, a=k, scale=theta)

plt.plot(x, y, label=f'Gamma(k={k}, Œ∏={theta})')
plt.title("Gamma Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

# üü¶ „Ç´„Ç§Ôºí‰πóÂàÜÂ∏É (Chi-square distribution)

* Ê®ôÊ∫ñÊ≠£Ë¶èÂàÜÂ∏É„ÅÆ‰∫å‰πóÂíå„ÅßÂæó„Çâ„Çå„ÇãÂàÜÂ∏É„ÄÇÁµ±Ë®àÁöÑ„Å™Ê§úÂÆö„ÇÑÂàÜÊï£ÂàÜÊûê„Åß„Çà„Åè‰Ωø„Çè„Çå„Çã„ÄÇ
* ÈÄ£Á∂ö

```python
# „Ç´„Ç§2‰πóÂàÜÂ∏ÉÔºàÁ¢∫ÁéáÂØÜÂ∫¶Èñ¢Êï∞Ôºâ„ÇíÊèèÁîª„Åô„Çã„Ç∑„É≥„Éó„É´„Å™‰æã
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# „Éë„É©„É°„Éº„Çø
df = 5 # Ëá™Áî±Â∫¶

# x ÁØÑÂõ≤Ôºà‰∏äÂÅ¥ 99.9%ÁÇπ„Åæ„Åß„ÇíÊèèÁîªÁØÑÂõ≤„Å´„Åô„ÇãÔºâ
x = np.linspace(0, chi2.ppf(0.999, df), 600)

# Á¢∫ÁéáÂØÜÂ∫¶Èñ¢Êï∞ÔºàPDFÔºâ
y = chi2.pdf(x, df)

plt.figure(figsize=(6, 4))
plt.plot(x, y, lw=2, label=f"df={df}")
plt.xlabel("x")
plt.ylabel("Á¢∫ÁéáÂØÜÂ∫¶")
plt.title("„Ç´„Ç§2‰πóÂàÜÂ∏ÉÔºàPDFÔºâ")
plt.grid(True, alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

# üü¶ tÂàÜÂ∏É (Student's t distribution)

* ÊØçÂàÜÊï£„ÅåÊú™Áü•„ÅÆ„Å®„Åç„Å´Ê®ôÊú¨Âπ≥Âùá„ÅÆÊé®Ê∏¨„Å´Áî®„ÅÑ„Çâ„Çå„ÇãÂàÜÂ∏É„ÄÇÊ≠£Ë¶èÂàÜÂ∏É„Å´‰ºº„Å¶„ÅÑ„Çã„ÅåË£æ„ÅåÂéö„ÅÑ„ÄÇ
* ÈÄ£Á∂ö

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# „Éë„É©„É°„Éº„Çø
df = 10 # Ëá™Áî±Â∫¶

# xËª∏„ÅÆÁØÑÂõ≤
x = np.linspace(-5, 5, 500)

# tÂàÜÂ∏É„ÅÆÁ¢∫ÁéáÂØÜÂ∫¶Èñ¢Êï∞
y = t.pdf(x, df)

# ÊèèÁîª
plt.plot(x, y, label=f'tÂàÜÂ∏É (Ëá™Áî±Â∫¶={df})')
plt.title("tÂàÜÂ∏É")
plt.xlabel("x")
plt.ylabel("Á¢∫ÁéáÂØÜÂ∫¶")
plt.legend()
plt.grid(True)
plt.show()
```

# üü¶ fÂàÜÂ∏É (F distribution)

* ‰∫å„Å§„ÅÆÁã¨Á´ã„Åó„Åü„Ç´„Ç§‰∫å‰πóÂàÜÂ∏É„ÅÆÊØî„Åã„ÇâÂæó„Çâ„Çå„ÇãÂàÜÂ∏É„ÄÇÂàÜÊï£ÂàÜÊûê (ANOVA) „Å™„Å©„ÅßÂà©Áî®„Åï„Çå„Çã„ÄÇ
* ÈÄ£Á∂ö

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# „Éë„É©„É°„Éº„Çø
dof1, dof2 = 5, 10 # ÂàÜÂ≠ê„ÅÆËá™Áî±Â∫¶„ÄÅÂàÜÊØç„ÅÆËá™Áî±Â∫¶

# x„ÅÆÁØÑÂõ≤
x = np.linspace(0, 5, 500)

# FÂàÜÂ∏É„ÅÆÁ¢∫ÁéáÂØÜÂ∫¶Èñ¢Êï∞
y = f.pdf(x, dof1, dof2)

# ÊèèÁîª
plt.plot(x, y, label=f"FÂàÜÂ∏É (df1={dof1}, df2={dof2})")
plt.title("FÂàÜÂ∏É")
plt.xlabel("x")
plt.ylabel("Á¢∫ÁéáÂØÜÂ∫¶")
plt.legend()
plt.grid(True)
plt.show()
```
